# INTRODUCTION TO EXPLORATORY DATA  ANALYSIS

Exploratory Data Analysis(EDA) is a process of examining the data, extracting insights of the data.Originally developed by American mathematician John Tukey in the 1970s, EDA techniques continue to be a widely used method in the data discovery process today.EDA is a crucial step in any data science project. It involves investigating and understanding a dataset to uncover patterns, trends, and potential issues. In statistics, EDA is an approach of analyzing data sets to summarize their main characteristics, often using statistical graphics and other data visualization methods.

### DISTPLOT

```python

plt.figure(figsize=(6,3.2),dpi=150)
plt.rc('xtick', labelsize=6)
plt.rc('ytick', labelsize=8)
sns.distplot(df['Unit_price'])
plt.show()
```

```python
plt.figure(figsize=(6,3.2),dpi=150)
plt.rc('xtick', labelsize=6)
plt.rc('ytick', labelsize=8)
sns.displot([df['Total'],df['cogs']])
plt.title('Total vs cogs', fontsize = 20)
plt.show()
```

### HISTPLOT

```python
sns.histplot(df['Total'],palette='hls')
plt.show()
```

### BOX PLOT

```python

sns.boxplot(x=df['City'],y=df['gross_income'])
plt.title('City vs gross_income', fontsize = 15)
plt.show()
```

### LINE PLOT

```python
sns.lineplot(x='Product_line',y='Unit_price',data=df,color='red')
plt.title('Product_line vs Unit_price', fontsize = 20)
plt.show()
```

### PIE CHART

```python
df['Payment'].value_counts().plot( kind='pie', y='Payment',autopct='%1.0f%%')
plt.legend(title='Payment_method')
plt.title("PAYMENT METHOD")
plt.show()
```

### BAR PLOT

```python
sns.barplot(x='Quantity',y='Product_line',data=random,color='red')
plt.show()
```

### SCATTER PLOT

```python
df1 = df[df['Total']>550]
x=df1['gross_income']
y=df1['Product_line']
plt.scatter(x, y,cmap='viridis')
plt.colorbar()

plt.show()
```

### HEATMAP

```python
a=sns.heatmap(df.corr(numeric_only=True))
a
```

### PAIR PLOT

```python
df1 = df[df['Total']>550]
sns.pairplot(df1)
plt.show()
```
